```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
sda = pd.read_csv('fifa_eda_stats.csv')
sda
```

```python
sda.shape
```

```python
sda.replace("€", "", regex=True, inplace=True)  # Remove currency symbol
sda.fillna(np.nan, inplace=True)  # Replace empty values with NaN
```

```python
# Histogram of player ages
plt.figure(figsize=(10, 6))
sns.histplot(data=sda, x='Age', bins=30, kde=True)
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of player Overall vs Potential
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sda, x='Overall', y='Potential', alpha=0.5)
plt.title('Overall vs Potential')
plt.xlabel('Overall')
plt.ylabel('Potential')
plt.show()
```

```python

# Bar plot of top 15 Nationalities
plt.figure(figsize=(12, 6))
top_nationalities = sda['Nationality'].value_counts().head(15)
sns.barplot(x=top_nationalities.index, y=top_nationalities.values)
plt.title('Top 10 Nationalities')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Example: Scatter plot between Overall Rating and Potential
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Overall', y='Potential', data=sda, hue='Preferred Foot')
plt.title('Overall Rating vs Potential by Preferred Foot')
plt.xlabel('Overall Rating')
plt.ylabel('Potential')
plt.legend(title='Preferred Foot')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Simulated data for player performance over seasons
# Create a DataFrame with columns: Season, Player, Performance
seasons = ['Season 1', 'Season 2', 'Season 3', 'Season 4']
players = ['Player A', 'Player B', 'Player C', 'Player D']
performance_data = []

for season in seasons:
    for player in players:
        performance_data.append({
            'Season': season,
            'Player': player,
            'Performance': random.randint(60, 100)  # Simulated performance score
        })

# Convert the data to a DataFrame
performance_df = pd.DataFrame(performance_data)

# Line plot to visualize player performance over seasons
plt.figure(figsize=(10, 6))
for player in players:
    player_data = performance_df[performance_df['Player'] == player]
    plt.plot(player_data['Season'], player_data['Performance'], marker='o', label=player)

plt.title('Player Performance Over Seasons')
plt.xlabel('Season')
plt.ylabel('Performance')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
